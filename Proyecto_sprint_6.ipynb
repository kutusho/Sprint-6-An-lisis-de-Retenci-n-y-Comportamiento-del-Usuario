Comentarios Iván: Deseo comenzar a atender los valores ausentes conforme lo siguiente:
name lo eliminaré porque si no se el nombre del juego difilmente será útil.
year_of_release con el valor (9999).
critic_score con el valor de la mediana por la cantidad de datos ausentes.
user_score con el valor del promedio porque me hace sentido dejar un valor que considere todos los criterios.
rating con la palabra Unknown.


```python
df_games = df_games.dropna(subset=['name'])
```

```python
df_games['year_of_release'] = df_games['year_of_release'].fillna(9999)
```

```python
df_games['critic_score'] = df_games['critic_score'].fillna(df_games['critic_score'].median())
```

```python
df_games['user_score'] = df_games['user_score'].fillna(df_games['user_score'].mean())
```

```python
df_games['rating'] = df_games['rating'].fillna('Unknown')
```

```python
df_games.isna().sum()
```

```python
df_games.info()
```

```python
df_games['total_sales'] = df_games[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)
```

```python
df_games['year_of_release'].value_counts().sort_index()
```

```python
df_games['year_of_release'].value_counts().sort_index().plot(kind='bar', figsize=(12, 5))
plt.title('Número de juegos lanzados por año')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Cantidad de juegos')
plt.show()

```

```python
plataforma_ano = df_games.pivot_table(
    index='year_of_release',
    columns='platform',
    values='name',
    aggfunc='count',
    fill_value=0
)
```

```python
plataforma_ano
```

```python
plataforma_ano.plot(kind='bar', figsize=(12, 5))
plt.title('Número de juegos lanzados por año')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Cantidad de juegos')
plt.show()
```
